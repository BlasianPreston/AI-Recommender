```python
marvel_movies = pd.read_csv('mcu_box_office.csv')
marvel_movies.head()
```

```python
matrix = marvel_movies.pivot_table(index='mcu_phase', columns='movie_title', values='tomato_meter')
```

```python
matrix = matrix.fillna(50, axis=0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(matrix.T)
```

```python
def get_recommendations(title, similarity=similarity,movies=marvel_movies):
  idx = marvel_movies[marvel_movies['movie_title'] == title].index[0]
  movie_scores = list(enumerate(similarity[idx]))
  sorted_scores = sorted(movie_scores, key=lambda x: x[1], reverse=True)
  top_scores = sorted_scores[1:11]
  movie_indices = [i[0] for i in top_scores]
  return marvel_movies['movie_title'].iloc[movie_indices]
```

```python
print(get_recommendations("Iron Man"))
```
